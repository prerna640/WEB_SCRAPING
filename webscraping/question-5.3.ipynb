# Top 10 ODI bowlers along with the records of their team and rating..

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/test")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
heds = soup.find_all('td',class_="table-body__cell name")
```

```python
links = []
for h in heds:
    a = h.find('a')
    links.append(a.text)
```

```python
links
```

```python
rating=[]
titles11=soup.find_all('span', class_="table-body__logo-text")
titles11[0:4]
```

```python
for i in titles11:
    rating.append(i.get_text().replace("\n",""))
rating
```

```python
rate=[]
titles111=soup.find_all('td', class_="table-body__cell u-text-right rating")
titles111[0:4]
```

```python
for i in titles111:
    rate.append(i.get_text().replace("\n",""))
rate
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['Player']=links[9:20]
jobs['Team']=rating[9:20]
jobs['rating']=rate[9:20]
```

```python
new_row = pd.DataFrame({'Player':'Meg Lanning','Team':'AUS', 'rating':749, },index =[0]) 
# simply concatenate both dataframes 
df = pd.concat([new_row,jobs]).reset_index(drop = True) 
df.head(8)
```
