```python
from bs4 import BeautifulSoup
import requests

#now , we will send get request to the webpages server to get the code of the page which is to be scraped by using the request library.
```

```python
page=requests.get("https://www.indeed.com/jobs?q=machine+learning&l=delhi")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

so, Now lets first create 4 empty lists . In these lists the data will be stored while scraping . We have created 4 empty lists for features which we have to exract.

1. Job_title
2.company_names
3. Location_list
4. Days_posted

```python
job_titles=[]
company_names=[]
locations_list=[]
#days_posted=[]
```

```python
#first, we will extract at the html tags where we have the job titles.Let me first show you on the web page in which tags the job title are put.
```

```python
# so lets extract all the tags having the job-titles
titles=soup.find_all('a', class_="jobtitle turnstileLink")
titles[0:4]
```

```python
# now the text of the job title is in "b" tags inside the tags extracted above.
# So we will run loop to iterate over the tags extracted above and one loop to iterate over the "b" tags inside it.
for i in titles:
    job_titles.append(i.get_text().replace("\n",""))
job_titles    
    
```

```python
titles[0].get_text()
```

```python
# so lets extract all the tags having the company names

companies=soup.find_all('span', class_="company")
companies
```

```python
for i in companies:
    company_names.append(i.get_text().replace("\n",""))                       
                         
```

```python
company_names
```

```python
#now we will extract all the html tags where we have how many days before them
```

```python
days=soup.find_all('span',class_="date")
days
```

```python
locations=soup.find_all('span',class_="location accessible-contrast-color-location")
locations[0:3]
```

```python
for i in locations:
    locations_list.append(i.text)
    
```

```python
locations_list
```

```python
print(len(job_titles),len(company_names),len(locations_list))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['company']=company_names
jobs['location']=locations_list
```
