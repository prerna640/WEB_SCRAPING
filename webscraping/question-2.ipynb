# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
heds = soup.find_all('h3',class_="lister-item-header")
```

```python
links = []
for h in heds:
    a = h.find('a')
    links.append(a.text)
```

```python
links
```

```python
titles1=[]
titles=soup.find_all('h3', class_="lister-item-header")
titles[0:4]
```

```python
for i in titles:
    
    titles1.append(i.get_text().replace("\n",""))
titles1
```

```python
year=[]
y=soup.find_all('span', class_="lister-item-year text-muted unbold")
y[0:4]
```

```python
for i in y:
    year.append(i.text)
year
```

```python
Rate=[]
y1=soup.find_all('div', class_="inline-block ratings-imdb-rating")
y1
```

#main > div > div.lister > div.lister-list > div> div.lister-item-content > div.ipl-rating-widget > div.ipl-rating-star.small > span.ipl-rating-star__rating

```python
for i in y1:
    Rate.append(i.text.replace("\n"," "))
Rate
```

```python
rate= Rate[100:200]
```

```python
rate
```

```python
len(year)
```

```python
print(len(rate),len(year),len(titles1))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['title']=links
jobs['year']=year
jobs['rate']=rate
```

```python
jobs
```

```python
jobs.to_csv("q2.csv", index= False)
```
